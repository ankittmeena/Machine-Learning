```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
data = pd.read_csv("USA_Housing.csv")
data.head()
```

```python
data = data.drop(['Address'],axis=1)
data=(data-data.min())/(data.max()-data.min())
data.head()
#normalize using min-max scaling
```

```python
Y = data['Price']
X = data.drop(['Price'],axis=1)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

```python
X_test = np.array(X_test)
X_train = np.array(X_train)
Y_test = np.array(Y_test)
Y_train = np.array(Y_train)
#numpy array is more efficient for large dataset computation
```

```python
m1 = len(X_test)
m2 = len(X_train)
m3 = len(X)
print(m1,m2,m3)
```

```python
def compute_cost(x, y, w, b): 
    m = len(x)
    total_cost = 0
    f = np.dot(w,x)+b
    total_cost = np.dot(f-y,f-y)
    total_cost = total_cost/(2*m)
    return total_cost
# to compute the cost function, given the model parameters w and b and the training data x and y.
```

```python
def compute_gradient(x, y, w, b): 
    m = len(x)
    dj_dw = 0
    dj_db = 0
    f = np.dot(w,x)+b
    dj_db = np.sum(f-y)
    dj_db = dj_db/m
    dj_dw = np.dot(f-y,x.T)
    dj_dw = dj_dw/m
    return dj_dw, dj_db
#gradient of the cost function
```

```python
def gradient_descent(x, y, w,b,alpha, num_iters):
    m = len(x)
    J_history = []
    for i in range(num_iters):
        dj_dw, dj_db = compute_gradient(x, y, w,b)
        w = w - alpha * dj_dw               
        b = b - alpha * dj_db             
        if i<100000:      
            cost =  compute_cost(x, y, w, b)
            J_history.append(cost)

        if i% math.ceil(num_iters/10) == 0:
            print(f"Iteration {i:4}: Cost {float(J_history[-1]):8.6f} ")
    
    return w,b
#Gradient descent to minimize the cost function
```

```python
w_in = np.array([0,0,0,0,0])

w,b = gradient_descent(X_train.T,Y_train.T,w_in,0.1,0.001,1000)
print("w,b found by gradient descent:",w,b)
```

```python
m = len(X_test)
predicted = np.zeros(m)
X_index = np.zeros(m)
for i in range(m):
    predicted[i] = np.dot(w,X_test[i]) + b
    X_index[i]= i
```

```python
plt.plot(X_index,predicted, c = "b")
plt.scatter(X_index,Y_test, marker='x', c='r') 
plt.title("House Price Prediction")
# Set the y-axis label
plt.ylabel('Price')
# Set the x-axis label
plt.xlabel('Features Index')
```
